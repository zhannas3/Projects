```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def logmodel(N,t,r):
    B = 80
    dNdt = r*(1-N/B)*N
    return dNdt
N0 = 7
t = np.linspace(0,60)

r= 0.09
N1 = odeint(logmodel,N0,t,args=(r,))
r = 0.15
N2 = odeint(logmodel,N0,t,args=(r,))
r = 0.25
N3 = odeint(logmodel,N0,t,args=(r,))
r = 0.45
N4 = odeint(logmodel,N0,t,args=(r,))

plt.plot(t,N1,'r--',label='r=0.09')
plt.plot(t,N2,'b--',label='r=0.15')
plt.plot(t,N3,'g--',label='r=0.25')
plt.plot(t,N4,'y--',label='r=0.6')
plt.xlabel('t')
plt.ylabel('N(t)')
plt.legend()
plt.grid()
plt.show()
```
