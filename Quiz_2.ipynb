```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def gmodel(N,t,k):
    dNdt = -k*N*np.log(N)
    return dNdt
N0 = 7
t = np.linspace(0,60)

k= 0.03
N1 = odeint(gmodel,N0,t,args=(k,))
k = 0.04
N2 = odeint(gmodel,N0,t,args=(k,))
k = 0.06
N3 = odeint(gmodel,N0,t,args=(k,))
k = 0.09
N4 = odeint(gmodel,N0,t,args=(k,))

plt.plot(t,N1,'r--',label='k=0.03')
plt.plot(t,N2,'b--',label='k=0.04')
plt.plot(t,N3,'g--',label='k=0.06')
plt.plot(t,N4,'y--',label='k=0.09')
plt.xlabel('t')
plt.ylabel('N(t)')
plt.legend()
plt.grid()
plt.show()
```
